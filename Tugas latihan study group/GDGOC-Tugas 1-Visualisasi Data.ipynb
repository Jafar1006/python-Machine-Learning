Download Data from kaggle
Iris Dataset

```python
import kagglehub
import pandas as pd
import os
import matplotlib.pyplot as plt
# Unduh versi terbaru
path_dataset = kagglehub.dataset_download("arshid/iris-flower-dataset")

# Temukan file CSV di dalam folder yang diunduh
for nama_file in os.listdir(path_dataset):
    if nama_file.endswith(".csv"):
        path_file_csv = os.path.join(path_dataset, nama_file)
        break

# Baca file CSV
Data_iris = pd.read_csv(path_file_csv)
Data_iris
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming 'species' is the column you want to create a pie chart for
species_counts = Data_iris['species'].value_counts()

# Create the pie chart
species_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90)

# Add a title
plt.title('Distribution of Iris Species')

# Ensure the circle's proportion
plt.axis('equal')

# Show the chart
plt.show()
```
